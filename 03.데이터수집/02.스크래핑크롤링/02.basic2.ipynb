* 브라우저 기능 컨트롤 

```python
import selenium 
from selenium import webdriver
# from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import  Service
import time

url = "http:/naver.com"

service = Service() # 크롬 드라이버 실행을 관리해 주는 객체 선언
options = webdriver.ChromeOptions()  # 브라우저 실행 환경 세팅(브라우저 실행 유무 결정, GPU가속 끄기, 창 크기 결정)
driver = webdriver.Chrome(service=service, options=options) # service, option설정을 한 객체(실행 : 브라우저 컨트롤 가능 객체)

driver.get(url)
```

```python
sec = 2
time.sleep(sec)
driver.back()
time.sleep(sec)
driver.forward()
time.sleep(sec)
driver.refresh()
time.sleep(sec)
driver.back()
time.sleep(sec)
driver.quit()
```

* 키 action처리

```python
from selenium import webdriver

from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

service = Service()
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)

url = "http://naver.com"
driver.get(url)

element = driver.find_element(By.ID, "query")
element.send_keys("컴퓨터")
element.send_keys(Keys.ENTER)
```

```python
try : 
    #네이버 이동 
    driver.get(url)
    element = driver.find_element(By.CLASS_NAME, "MyView-module__link_login___HpHMW")
    element.click()
    
    #id, pass입력
    
    driver.find_element(By.ID, "id").send_keys("su0905jin")
    driver.find_element(By.ID, "pw").send_keys("imsujin0905!")
    
    driver.find_element(By.ID, "log.login").click()
    print(driver.page_source)
    
except Exception as e: 
    print(e)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pyperclip 

url = "http://naver.com"
service = Service()
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)
driver.get(url)

try : 
    #네이버 이동 
    driver.get(url)
    
    #로그인 버튼 클릭 
    element = driver.find_element(By.CLASS_NAME, "MyView-module__link_login___HpHMW")
    element.click()
    
    #id, pass입력 
    pyperclip.copy("su0905jin")
    driver.find_element(By.ID, "id").send_keys(Keys.CONTROL,"v")
    
    pyperclip.copy("imsujin0905!")
    driver.find_element(By.ID, "pw").send_keys(Keys.CONTROL,"v")
    
    driver.find_element(By.ID, "log.login").click()
    print(driver.page_source)
    
except Exception as e: 
    print(e)
```

```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import  Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import requests 

service = Service()
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)


#for year in range(2020,2025): 
url = "https://search.daum.net/search?w=tot&q=2024%EB%85%84%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84&DA=MOR&rtmaxcoll=MOR"
driver.get(url)
    
main = driver.find_element(By.CLASS_NAME,"pdt2")
lis = main.find_elements(By.TAG_NAME,"li")

   
for idx, li in enumerate(lis):
    title = li.find_element(By.CLASS_NAME,"tit-g").text
    contents = li.find_element(By.CLASS_NAME, "item-contents").text
    img_url = li.find_element(By.TAG_NAME,"img").get_attribute("src")


    img_res = requests.get(img_url)
    img_res.raise_for_status()
    
    
    with open("./images-daum/image{0}.jpg".format(idx + 1), "wb") as f : 
        f.write(img_res.content)
    print(title)
    print(contents)
    print(img_url)
```
