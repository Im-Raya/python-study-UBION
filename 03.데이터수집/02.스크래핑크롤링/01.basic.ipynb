```python
# 크롤링
import requests
from bs4 import BeautifulSoup


url = f"https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&qvt=0&query=2024%EB%85%84%20%EC%98%81%ED%99%94%20%EC%88%9C%EC%9C%84"
res = requests.get(url)
res.raise_for_status()

print(len(res.text))
```

```python

#스크래핑 
soup = BeautifulSoup(res.text, "html.parser")

movies = soup.find("div", attrs= {"class" : "movie_audience_ranking _main_panel"})
movies1 = movies.find("div")
movie_lis = movies1.find_all("li")
movie_lis
```

```python
for idx,  movie in enumerate(movie_lis):
    title= movie.find("strong", attrs={"class" : "scm_ellipsis_text _text"}).text.strip()
    items = movie.find("dl", attrs={"class" : "movie_item"}).text.strip()
    img = movie.find("img") #이미지 불러옴
    img_url = img['src'] #이미지 주소만 
    
    print(title,":", items, img_url)
    print('-'*50)
```
