```python
# 한국 지수 예시(코스피, 코스닥)
import yfinance as yf
import matplotlib.pyplot as plt

#코스피 지수 (KS11)
kospi = yf.download('^KS11', start = '2023-01-01', end='2025-08-28')

#코스닥 지수 (KQ11)
kosdaq = yf.download('^KS11', start = '2023-01-01', end='2025-08-28')

plt.figure(figsize=(10,6))
plt.plot(kospi['Close'], label='KOSPI')
plt.plot(kosdaq['Close'], label='KOSDAQ')
plt.title('KOSPI & KOSDAQ 2023')
plt.legend()
plt.show()

```
