* 항아리 그래프

* 구현 순서
    * 남성 데이터 리스트m에 저장
    * 여성 데이터 리스트f에 저장
    * reverse() 함수를 이용해서 f리스트 값을 역순으로 재배열

```python
import csv
f = open('./data/gender(1).csv', 'r', encoding='cp949')
data = csv.reader(f)

m=[]
f=[]

for row in data: 
    if '신도림' in row[0]:
        for i in range(0,101):
            m.append(-int(row[i+3])) #인구 피라미드를 그릴 때 양쪽으로 나뉘어 보이도록 하기 위해
            f.append(int(row[-(i+1)])) #남성 데이터를 음수(-)
            
f.reverse()
plt.rc('font',family='malgun gothic')
plt.title('신도림 지역의 남녀 성별 인구 분포')
plt.rcParams['axes.unicode_minus'] = False
plt.barh(range(101), m, label='남성')
plt.barh(range(101), f, label='여성')
plt.legend()
plt.show()

```

```python
# 제주특별자치도 
# 그래프 데이터에 던지는 질문: 제주도에는 어떤 성별이 더 많이 살고 있는가?

import csv 
import matplotlib.pyplot as plt

f1 = open(r'./data/gender(1).csv', encoding= 'cp949')
data = csv.reader(f1)
m= []
f= []

# name = input('찾고 싶은 지역의 이름을 알려주세요: ')

name = input('찾고 싶은 지역의 이름을 알려주세요: ') #'제주특별자치도' #test code
for row in data : 
    if name in row[0] : 
        for i in row[3:104] : 
            m.append(int(i))
        for i in row[106:] :
            f.append(int(i))
        break 
    
plt.rc('font', family= 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('특정 지역의 남녀 성별 인구 분포')
plt.barh(range(101),  m, label = '남성')
plt.barh(range(101),  f, label = '여성')
plt.legend()
plt.show()
```

# 컬러 테이블

```python
import csv

f= open('./data/gender(1).csv' ,'r',encoding='cp949')
data = csv.reader(f)
size = []

name = input('찾고 싶은 지역의 이름을 알려주세요 : ')

for row in data :
    if name in row[0] :
        m = 0
        f = 0
        for i in range(101) :
            m += int(row[i+3])
            f += int(row[i+106])
        break
size.append(m)
size.append(f)

import matplotlib.pyplot as plt

plt.rc('font', family ='Malgun Gothic')
color = ['crimson', 'darkcyan']
plt.axis('equal')
plt.pie(size, labels = ['남','여'], autopct ='%.1f%%', colors = color, startangle =90)
plt.title(name + ' 지역의 남녀 성별 비율')
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import colors as mcolors
colors = dict(mcolors.BASE_COLORS, **mcolors.CSS4_COLORS)
# Sort colors by hue, saturation, value and name.
by_hsv = sorted((tuple(mcolors.rgb_to_hsv(mcolors.to_rgba(color)[:3])), name)
                for name, color in colors.items())
sorted_names = [name for hsv, name in by_hsv]
n = len(sorted_names)
ncols = 4
nrows = n // ncols
fig, ax = plt.subplots(figsize=(9, 8), dpi = 300)
# Get height and width
X, Y = fig.get_dpi() * fig.get_size_inches()
h = Y / (nrows + 1)
w = X / ncols
for i, name in enumerate(sorted_names):
    row = i % nrows
    col = i // nrows
    y = Y - (row * h) - h
    xi_line = w * (col + 0.05)
    xf_line = w * (col + 0.25)
    xi_text = w * (col + 0.3)
    ax.text(xi_text, y, name, fontsize=(10),
            horizontalalignment='left',
            verticalalignment='center')
    ax.hlines(y + h * 0.1, xi_line, xf_line,
              color=colors[name], linewidth=(6))
ax.set_xlim(0, X)
ax.set_ylim(0, Y)
ax.set_axis_off()
fig.subplots_adjust(left=0, right=1,
                    top=1, bottom=0,
                    hspace=0, wspace=0)
plt.show()
```
